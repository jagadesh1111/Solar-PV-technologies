```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("/content/01Feb2016-CR3000.csv")
df.head()
```

```python
df.drop(['DL GR1','DL GR2','DL Panel_T1','DL Panel_T2','DL Panel_T3','DL Panel_T4','DL Panel_T5'],axis=1)
#dropping the unnecessary columns
```

```python
from sklearn.linear_model import LinearRegression
x1 = df['DL Bat. V'].values.reshape(-1, 1)
y1= df['DL WS_ms'].values.reshape(-1, 1)
#Radiation vs  Wind speed
```

```python
model = LinearRegression()
model.fit(x1, y1)
```

```python
model = LinearRegression().fit(x1, y1)
r_sq = model.score(x1, y1)
print(f"coefficient of determination: {r_sq}")
```

```python
plt.scatter(x1,y1)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
x2 = df['DL Bat. V'].values.reshape(-1, 1)
y2= df['DL RTD_Airtemp'].values.reshape(-1, 1)
#Radiation vs  Air temperature
```

```python
model = LinearRegression()
model.fit(x2, y2)
model = LinearRegression().fit(x2, y2)
r_sq = model.score(x2, y2)
print(f"coefficient of determination: {r_sq}")
```

```python
plt.scatter(x2,y2)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
x3 = df['DL Bat. V'].values.reshape(-1, 1)
y3= df['DL WindDir'].values.reshape(-1, 1)
#Radiation vs  Wind direction
```

```python
model = LinearRegression()
model.fit(x3, y3)
model = LinearRegression().fit(x3, y3)
r_sq = model.score(x3, y3)
print(f"coefficient of determination: {r_sq}")
```

```python
plt.scatter(x3,y3)
plt.show()
```
